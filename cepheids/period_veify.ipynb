We already plotted the RV curves. We need to verify the period lies within periods determined by Storm 2011 et al. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vel_data = pd.read_csv("E:/Projects/hubble_constant_project/hubble_const_data/miras_data/Storm_2011_vel_timeseries.csv")
# vel_data = vel_data.loc[:]
full_dist_mag_data = pd.read_csv("E:/Projects/hubble_constant_project/hubble_const_data/miras_data/Storm_2011_dist_mag.csv")
# full_dist_mag_data = full_dist_mag_data.loc[:]
```

```python
object_name = 'HV6093 '  
filtered_data = vel_data.loc[vel_data['Name'] == object_name]

plt.plot(filtered_data['Heliocentric Julian Date'], filtered_data['Heliocentric Radial Velocity(km/s)'], marker='o', linestyle='-')
plt.title(f'Radial Velocity for {object_name}')
plt.xlabel('Heliocentric Julian Date')
plt.ylabel('Heliocentric Radial Velocity (km/s)')
plt.axvline(x=2.8 + 2.4537e6, color='red', linestyle='--', linewidth=1)
plt.axvline(x=7.8 + 2.4537e6, color='red', linestyle='--', linewidth=1)

print(f'Obtained log(period) = {np.log10(7.8-2.8)} days')
print('Actual log(period) by Storm et al = ' + str(full_dist_mag_data.loc[full_dist_mag_data['Name'] == object_name]['logP'].values[0]) + ' days')

plt.grid()
plt.show()
```
