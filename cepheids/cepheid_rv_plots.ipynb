```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vel_data = pd.read_csv("E:/Projects/hubble_constant_project/hubble_const_data/miras_data/Storm_2011_vel_timeseries.csv")
vel_data = vel_data.loc[:]
full_dist_mag_data = pd.read_csv("E:/Projects/hubble_constant_project/hubble_const_data/miras_data/Storm_2011_dist_mag.csv")
full_dist_mag_data = full_dist_mag_data.loc[:]
```

```python
dist_mag_data = full_dist_mag_data[full_dist_mag_data['Radvel_data_available'] == 'Y'] # Selecting only those objects for which Radvel data is available
```

There are a few objects in 'full_dist_mag_data' not included in 'vel_data'. So we will seperate them and work only on the rest of the data.

```python
names = dist_mag_data['Name']

for name in names:                                     # to plot RV curve for each object in data

    vel_of_object = vel_data[vel_data['Name'] == name]
    hjd = vel_of_object['Heliocentric Julian Date']
    hrv = vel_of_object['Heliocentric Radial Velocity(km/s)']

    hjd_arr = np.array(hjd)
    hrv_arr = np.array(hrv)         # These two form arrays of each object's HJD and RV during each cycle
    
    print(f'RV Phase plot for object {name}')
    plt.plot(hjd_arr, hrv_arr, marker='o', linestyle='-')  
    plt.xlabel('HJD')
    plt.ylabel('Heliocentric Velocity (Km/s)')

    plt.show()
```

Some of the RV curves are distorted, mostly due to the observational errors, or phase distribution errors. We will only focus on the clean RV curves to avoid any bias in the final result. We will first verify the periods one by one, and use them for further analysis.
