```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
vel_data = pd.read_csv("E:/Projects/hubble_constant_project/hubble_const_data/miras_data/Storm_2011_vel_timeseries.csv")
vel_data = vel_data.loc[:]
full_dist_mag_data = pd.read_csv("E:/Projects/hubble_constant_project/hubble_const_data/miras_data/Storm_2011_dist_mag.csv")
full_dist_mag_data = full_dist_mag_data.loc[:]
```

```python
log_P = full_dist_mag_data['logP']
dist_modulus = full_dist_mag_data['(m-M)0']
m = full_dist_mag_data['VMAG']
```

```python
result = stats.linregress(log_P, m)  
a = result.slope
b = result.intercept 
print(a,b)

M_obs = a*log_P + b

plt.plot(log_P, m, 'o', label='original data')
plt.plot(log_P, a*log_P + b, 'r', label='fitted line')
plt.title('PLR')
plt.xlabel('log(P)')
plt.ylabel('Apparent Magnitude')
plt.legend()
```

```python
log_H0 = (M_obs + 5*a +25)/5
H0 = 10**log_H0[0:4]
print(f'Hubble constant values in km s^-1 Mpc^-1 are:  ')
H0
```

These Hubble constant values are significantly lower than that of type 1 SN, mostly because of the small sample size. The SN data had nearly 800 SN's at scattered redshifts which were analysed to find H0. In the SN analysis, we obtained that redshifts between
a range 0.05 < z < 0.1 give values matching with current literature. 

Whereas here, we have only 36 objects, incorporating a significant bias. In addition, the data only contains redshifts significantly below z=0.1. This results in a deviation from the currently predicted value of H0.

But the value agrees with the previously agreed values of H0, ranging from 50 to 100, based on different distance ladders used. This historical discrepancy is well explained [Here](https://iopscience.iop.org/article/10.1086/340057/fulltext/15433.text.html) & [Here](https://lweb.cfa.harvard.edu/~dfabricant/huchra/hubble/).
